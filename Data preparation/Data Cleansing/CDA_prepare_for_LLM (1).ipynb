# Sampling

```python
def parse_dates_by_portal(df, date_col="DATUM", portal_col="PORTAL"):
    df = df.copy()

    df[portal_col] = df[portal_col].astype(str).str.strip()

    raw_dates = (
        df[date_col]
        .astype(str)
        .str.strip()
        .str.replace("\ufeff", "", regex=False)
        .str.replace(r"\s+", " ", regex=True)
    )

    parsed = pd.Series(pd.NaT, index=df.index, dtype="datetime64[ns]")

    portal_formats = {
        # Klix examples: 2023-08-20, 2019-01-10
        "Klix.ba": [
            "%Y-%m-%d",
            "%Y-%m-%d %H:%M:%S",
        ],

        # RTRS examples: 04/07/2021 means 4 July 2021
        "RTRS": [
            "%d/%m/%Y",
            "%d/%m/%Y %H:%M",
            "%d/%m/%Y %H:%M:%S",
        ],

        # Usually Bosnian portal style
        "Radio Sarajevo": [
            "%d.%m.%Y",
            "%d.%m.%Y.",
            "%d.%m.%Y %H:%M",
            "%d/%m/%Y",
        ],

        "6yka": [
            "%d.%m.%Y",
            "%d.%m.%Y.",
            "%d/%m/%Y",
            "%Y-%m-%d",
        ],

        "Dnevni Avaz": [
            "%d.%m.%Y",
            "%d.%m.%Y.",
            "%d/%m/%Y",
            "%Y-%m-%d",
        ],

        "STAV": [
            "%d.%m.%Y",
            "%d.%m.%Y.",
            "%d/%m/%Y",
            "%Y-%m-%d",
        ],
    }

    for portal, formats in portal_formats.items():
        portal_mask = df[portal_col].eq(portal)

        for fmt in formats:
            idx = portal_mask & parsed.isna()

            if idx.sum() == 0:
                break

            parsed.loc[idx] = pd.to_datetime(
                raw_dates.loc[idx],
                format=fmt,
                errors="coerce"
            )

    # Fallback za sve što nije prošlo gore
    # dayfirst=True jer većina BiH portala koristi dan/mjesec/godina
    idx = parsed.isna()

    if idx.sum() > 0:
        parsed.loc[idx] = pd.to_datetime(
            raw_dates.loc[idx],
            errors="coerce",
            dayfirst=True
        )

    return parsed
```

```python
N_SAMPLE = 10000
RANDOM_STATE = 43

THEMES = {
    "euroatlantske_integracije": {
        "score": "euroatlantske_integracije_score",
        "overlap": "euroatlantske_integracije_overlap",
        "matched": "euroatlantske_integracije_matched",
    },
    "negiranje_genocida": {
        "score": "negiranje_genocida_score",
        "overlap": "negiranje_genocida_overlap",
        "matched": "negiranje_genocida_matched",
    },
    "gradjanska_vs_konstitutivni": {
        "score": "gradjanska_vs_konstitutivni_score",
        "overlap": "gradjanska_vs_konstitutivni_overlap",
        "matched": "gradjanska_vs_konstitutivni_matched",
    },
    "izborna_reforma": {
        "score": "izborna_reforma_score",
        "overlap": "izborna_reforma_overlap",
        "matched": "izborna_reforma_matched",
    },
}


# =========================
# HELPER FUNKCIJE
# =========================

def is_non_empty_matched(x):
    """
    Provjerava da li matched kolona stvarno ima neki keyword.
    Radi i za NaN, "", "[]", "{}", "nan", "None".
    """
    if pd.isna(x):
        return False

    s = str(x).strip().lower()

    if s in ["", "[]", "{}", "nan", "none", "null"]:
        return False

    return True


def allocate_balanced_quotas(counts, total_n):
    """
    Pravi balansirane kvote po stratumu.
    Stratum = period_7d + portal + tema.

    Ako neki stratum ima malo članaka, uzme koliko može,
    a ostatak raspodijeli na druge stratume.
    """
    counts = counts.astype(int)
    total_available = counts.sum()

    if total_available == 0:
        raise ValueError("Nema dostupnih članaka za sampling.")

    total_n = min(total_n, total_available)

    quotas = pd.Series(0, index=counts.index, dtype=int)
    remaining_capacity = counts.copy()

    remaining_n = total_n

    while remaining_n > 0:
        active = remaining_capacity[remaining_capacity > 0]

        if len(active) == 0:
            break

        # Ako je ostalo manje uzoraka nego aktivnih grupa,
        # dodijeli po 1 grupama s najviše preostalih članaka.
        if remaining_n < len(active):
            chosen = active.sort_values(ascending=False).index[:remaining_n]
            quotas.loc[chosen] += 1
            break

        base_quota = remaining_n // len(active)

        add = active.clip(upper=base_quota).astype(int)

        quotas.loc[add.index] += add
        remaining_capacity.loc[add.index] -= add

        remaining_n -= add.sum()

    return quotas
```

```python
import numpy as np
```

```python
df["DATUM"] = pd.to_datetime(df["DATUM"], errors="coerce")
df = df.dropna(subset=["DATUM", "PORTAL"]).copy()

df["PORTAL"] = df["PORTAL"].astype(str).str.strip()

if "article_id" not in df.columns:
    df["article_id"] = np.arange(len(df))

print("Originalna distribucija po portalima:")
print(df["PORTAL"].value_counts())


# Napravi ID članka ako ne postoji
if "article_id" not in df.columns:
    df["article_id"] = np.arange(len(df))

# =========================
# 7-DNEVNE GRUPE
# =========================

PERIOD_DAYS = 15

min_date = df["DATUM"].min().normalize()

df["period_id"] = ((df["DATUM"] - min_date).dt.days // PERIOD_DAYS).astype(int)
df["period_start"] = min_date + pd.to_timedelta(df["period_id"] * PERIOD_DAYS, unit="D")
df["period_end"] = df["period_start"] + pd.Timedelta(days=PERIOD_DAYS - 1)
```

```python
theme_rows = []

for theme_name, cols in THEMES.items():
    score_col = cols["score"]
    overlap_col = cols["overlap"]
    matched_col = cols["matched"]

    # Ako neka kolona ne postoji, napravi default
    if score_col not in df.columns:
        df[score_col] = 0

    if overlap_col not in df.columns:
        df[overlap_col] = 0

    if matched_col not in df.columns:
        df[matched_col] = ""

    score = pd.to_numeric(df[score_col], errors="coerce").fillna(0)
    overlap = pd.to_numeric(df[overlap_col], errors="coerce").fillna(0)
    matched = df[matched_col].apply(is_non_empty_matched)

    # Članak pripada temi ako ima score, overlap ili matched keyword
    mask = (score > 0) | (overlap > 0) | matched

    temp = df.loc[mask].copy()
    temp["tema"] = theme_name
    temp["tema_score"] = score.loc[mask]
    temp["tema_overlap"] = overlap.loc[mask]
    temp["tema_matched"] = df.loc[mask, matched_col]

    theme_rows.append(temp)

if not theme_rows:
    raise ValueError("Nijedan članak nije pronađen ni za jednu temu.")

long_df = pd.concat(theme_rows, ignore_index=True)
```

```python
long_df = long_df.sort_values(
    by=["article_id", "tema_score", "tema_overlap"],
    ascending=[True, False, False]
)

long_df = long_df.drop_duplicates(subset=["article_id"], keep="first").copy()

# =========================
# STRATIFIKOVANI SAMPLING
# =========================

STRATA_COLS = ["period_start", "PORTAL", "tema"]

counts = long_df.groupby(STRATA_COLS).size()

quotas = allocate_balanced_quotas(counts, N_SAMPLE)
quotas = quotas[quotas > 0]

quota_df = quotas.rename("quota").reset_index()

sample_pool = long_df.merge(
    quota_df,
    on=STRATA_COLS,
    how="inner"
)

sampled_parts = []

for keys, group in sample_pool.groupby(STRATA_COLS, observed=True):
    q = int(group["quota"].iloc[0])

    sampled_group = group.sample(
        n=q,
        replace=False,
        random_state=RANDOM_STATE
    )

    sampled_parts.append(sampled_group)

sampled_df = pd.concat(sampled_parts, ignore_index=True)

# Shuffle finalni dataset
sampled_df = sampled_df.sample(
    frac=1,
    random_state=RANDOM_STATE
).reset_index(drop=True)

# Makni helper quota kolonu
sampled_df = sampled_df.drop(columns=["quota"], errors="ignore")
```
