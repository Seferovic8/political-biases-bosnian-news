# Radio Sarajevo

```python
import re
import pandas as pd


def parse_radio_sarajevo_txt(putanja: str, encoding: str = "utf-8") -> pd.DataFrame:
    """
    Učitava .txt fajl sa diska (po putanji) i pretvara članke u njemu
    (odvojene oznakom '<***>', sa poljima u obliku 'KLJUC: VRIJEDNOST')
    u pandas DataFrame.

    Parametri
    ---------
    putanja : str
        Putanja do .txt fajla. U Google Colabu, ako je fajl uploadovan,
        obično je to npr. "/content/clanci.txt" ili
        "/content/drive/MyDrive/.../clanci.txt" ako je u Google Drive-u.
    encoding : str
        Kodiranje fajla (default "utf-8").

    Vraća
    -----
    pd.DataFrame
        Kolone: NOVINA, DATUM, RUBRIKA, NADNASLOV, NASLOV, PODNASLOV,
                STRANA, AUTORI, TEKST
    """
    # Učitaj sadržaj fajla
    with open(putanja, encoding=encoding) as f:
        tekst = f.read()

    # Ukloni BOM (ako postoji) i podijeli na blokove po separatoru <***>
    tekst = tekst.lstrip("\ufeff")
    blokovi = [b.strip() for b in tekst.split("<***>") if b.strip()]

    # Polja koja očekujemo na početku svakog bloka (redoslijed je fiksan)
    polja = ["NOVINA", "DATUM", "RUBRIKA", "NADNASLOV", "NASLOV",
             "PODNASLOV", "STRANA", "AUTOR(I)"]

    zapisi = []

    for blok in blokovi:
        red = {p: None for p in polja}
        red["TEKST"] = ""

        linije = blok.splitlines()
        ostatak_idx = 0  # gdje počinje tekst članka

        for i, linija in enumerate(linije):
            linija_strip = linija.strip()

            if linija_strip == "":
                # Prazan red unutar zaglavlja - samo preskoči, ne znači
                # da je tekst članka počeo
                continue

            spojeno = False
            for p in polja:
                prefiks = f"{p}:"
                if linija_strip.startswith(prefiks):
                    vrijednost = linija_strip[len(prefiks):].strip()
                    red[p] = None if vrijednost == "N/A" else vrijednost
                    ostatak_idx = i + 1
                    spojeno = True
                    break
            if not spojeno:
                # Čim naiđemo na liniju koja nije meta-polje, prestajemo
                # tražiti dalja polja (ostatak je tekst članka)
                break

        # Sve od ostatak_idx nadalje je tekst članka
        red["TEKST"] = "\n".join(linije[ostatak_idx:]).strip()

        zapisi.append(red)

    df = pd.DataFrame(zapisi)
    df = df.rename(columns={"AUTOR(I)": "AUTORI"})
    df = df.rename(columns={"NOVINA": "PORTAL"})

    # Lijep redoslijed kolona
    kolone = ["PORTAL", "DATUM", "NASLOV",
                "TEKST"]
    df = df[kolone]

    return df
```

```python
import glob
import os

base = "drive/MyDrive/BHNC_Dataset/Radio Sarajevo"

folders = ["BiH", "Kolumne", "Metromahala - teme", "Region"]

txt_files = []

for folder in folders:
    pattern = os.path.join(base, folder, "*.txt")
    txt_files.extend(glob.glob(pattern))

print(len(txt_files))
txt_files[:5]
```

```python
df_list = [parse_radio_sarajevo_txt(fi) for fi in txt_files]
radio = pd.concat(df_list, ignore_index=True)
```

```python
radio = radio.dropna()
```
